```python
import numpy as np
import pandas as pd
from astropy import constants as ct
import matplotlib.pyplot as plt
from astropy_healpix import HEALPix
from astropy import units as u
```

```python
hpx = HEALPix(nside=512, order='ring')
```

```python
i=0    
spi=[]


"""FINDING PIXEL AT LAT=31.9505 deg SOUTH  and LONGITUDE STARTING FROM 0 AND INCREMENTING BY 15 Deg"""
while i <360:
    
    spi.append(hpx.lonlat_to_healpix(i*u.deg,-31.9505*u.deg))
    
    i+=15

    
"""Storing the pixel as numpy array"""
sat_pix_indices=np.array(spi) 
```

```python
R=6378.1    #Radius of earth in kms

r=np.logspace(np.log10(400),np.log10(36000),10) 


"""PERCENTAGE OF AREA SEEN BY SATELLITE     """
"""% AREA SEEN =   50/(1+ (R/d))   """

p_area=np.zeros(10)

for i in range(0,len(r)):
    p_area[i]= 50 / (1+ (R/r[i]))
    
    
for i in range(0,len(r)):
    print("AREA SEEN AT",r[i],"KMS IS ",p_area[i],"%")    
```

```python
hpix=data["HEALPIX INDICES"]

```

```python
"""For every pixel the satellite will cross here I am checking if the healpix indices are matching 
                 with the healpix indices of the transmitters,
If they are then the powers are added if not then the power remains the same"""


power=np.zeros(len(hpix))
for i in range(len(sat_pix_indices)):
    for j in range (len(hpix)):
        if sat_pix_indices[i]== hpix[j]:
            power[j]+=data.iloc[j]['Power at 400km']
        else:
            power[j]=data.iloc[j]['Power at 400km']
            
```

```python
data["pix power 400"]=power
```
